# **Calculadora Simples**

```python
# Inicia um bloco try-except para lidar com possíveis erros de entrada
try:
    # Solicita ao usuário o primeiro número e o converte para float
    num1 = float(input("Digite o primeiro número: "))
    # Solicita ao usuário o segundo número e o converte para float
    num2 = float(input("Digite o segundo número: "))
    # Solicita a operação desejada (somar ou subtrair) e converte para minúsculas
    operacao = input("Escolha a operação (somar ou subtrair): ").lower()

    # Verifica se a operação escolhida é 'somar'
    if operacao == "somar":
        # Realiza a soma dos dois números
        resultado = num1 + num2
        # Imprime o resultado da soma
        print(f"A soma é: {resultado}")
    # Verifica se a operação escolhida é 'subtrair'
    elif operacao == "subtrair":
        # Realiza a subtração do segundo número do primeiro
        resultado = num1 - num2
        # Imprime o resultado da subtração
        print(f"A subtração é: {resultado}")
    # Se a operação não for 'somar' nem 'subtrair'
    else:
        # Imprime uma mensagem de operação inválida
        print("Operação inválida. Por favor, escolha 'somar' ou 'subtrair'.")
# Captura o erro ValueError, que ocorre se a entrada não for um número
except ValueError:
    # Imprime uma mensagem de erro para entrada inválida
    print("Entrada inválida. Por favor, digite apenas números.")
```
